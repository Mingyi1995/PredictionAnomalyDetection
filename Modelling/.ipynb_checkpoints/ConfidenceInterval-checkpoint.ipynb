```python
import pandas as pd
import numpy as np
```

```python
def loadData(file):
    data = pd.read_csv(file)
    print('Raw shape: ',data.shape)
#     data['Date'] = pd.to_datetime(data.Date)
    print('Days: ',len(set(data.Date)))
    return data
```

```python
def getTimeSeries(df):
    table = pd.pivot_table(df, values='vehicle_count', index=['Date'],
                    columns=['DOLocationID'], aggfunc=np.sum, fill_value=0)

    return table
```

```python
hub = 'Penn'
rootDir = '/home/mingyi/Dropbox/UrbanTemporalNetworks/'
dataDir = rootDir + 'processedData/'

file = dataDir + hub + 'VehicleByHour.csv'
```

```python
rawdata = loadData(file)
edge_data = getTimeSeries(rawdata)
```

```python
edge_data
```

```python
# stdDF = pd.DataFrame()
for col in edge_data.columns:
    std = edge_data[col].std()
    edge_data[col] = edge_data[col].apply(lambda x: str(0)+';'+str(round(x+1.645*std,3)) 
                                          if (x-1.645*std)<0 
                                          else 
                                          str(round(x-1.645*std,3))+';'+str(round(x+1.645*std,3)))
```

```python
edge_data.to_csv('../processedData/%s90CI.csv'%hub)
```
