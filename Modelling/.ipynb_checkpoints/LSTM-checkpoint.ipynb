```python
import pandas as pd
import numpy as np
# from sklearn.metrics import mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
hub = 'LGA'
dataDir = '/home/mingyi/Dropbox/UrbanTemporalNetworks/processedData/'
file = dataDir + hub + 'VehicleByHour2018fromHub.csv'

externalDataDir = "/home/mingyi/Dropbox/UrbanTemporalNetworks/HongData/"+hub+'2018/'
extFile = externalDataDir+"external.csv"
```

```python
df = pd.read_csv(file)
df = df.groupby(['Date','Hour']).agg({'vehicle_count':'sum'}).reset_index()
external = pd.read_csv(extFile)
df = df.merge(external,on=['Date','Hour'])
df = df.drop(columns=['Date','Hour'])
df['lag'] = df['vehicle_count'].shift(1)
df = df.dropna()
# # normalization is a must in NN models
scaler = MinMaxScaler(feature_range = (0, 1))
```

```python
df
```

```python
# Still, training dataset will be from Jan 28th to Oct, testing dataset will be Nov and Dec

for col in df.columns:
    df[col] = df[col].astype(int)
features_set = df.drop(columns=['vehicle_count'])[:-31*24]

labels = df[['vehicle_count']][:-31*24]
features_set, labels = np.array(features_set), np.array(labels)
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units=7*24, return_sequences=True, input_shape=(features_set.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=7*24, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=7*24, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units = 1))
model.compile(optimizer = 'adam', loss = 'mean_absolute_error')

model.fit(features_set, labels, epochs = 50, batch_size = 32)
```

```python
model.summary()
```

### test

```python

features_set_test = df.drop(columns=['vehicle_count'])[-31*24:]
labels_test = df[['vehicle_count']][-31*24:]

features_set_test, labels_test = np.array(features_set_test), np.array(labels_test)
features_set_test = np.reshape(features_set_test, (features_set_test.shape[0], features_set_test.shape[1], 1))

predictions = model.predict(features_set_test)

print('out of sample r2',r2_score(labels_test,predictions))
```

```python
# mape could be probmatic as is may lead to division by 0
# add 1 to both residual and real value to prevent infinite values
insample = model.predict(features_set)

print('insample r2',r2_score(labels,insample))
```
